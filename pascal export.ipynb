```python
from xml.dom.minidom import parseString
from lxml.etree import Element, SubElement, tostring
import csv
```

```python

text_file_path = "grl1_yolo.txt"
f = open(text_file_path,"r")
line = f.readline()
detections={}
comps = []
while(line):
    
    line = line.replace("\n", "")
    comps = line.split(",")
    print(comps)
    if(comps[0] in detections):
        detections[comps[0]].append(list(map(float, comps[2:])))
    else:
        detections[comps[0]]=[]
        detections[comps[0]].append(list(map(float, comps[2:])))
    line=f.readline()
f.close()
print(detections)
classes = ['background',
           'aeroplane', 'bicycle', 'bird', 'boat',
           'bottle', 'bus', 'car', 'cat',
           'chair', 'cow', 'diningtable', 'dog',
           'horse', 'motorbike', 'person', 'pottedplant',
           'sheep', 'sofa', 'train', 'tvmonitor']
```

```python

import os
import csv
with open('gr_graal1.csv', 'w', newline='') as csvFile:
    writer = csv.writer(csvFile)
    tot = 1
    for k in detections:
        objs = detections[k]
        i=0
        tot+=1
        i
        for obj in objs:

            row=[]
            row.append(k.split('/')[1])
            statinfo = os.stat('../%s'%k)
            
            row.append(str(statinfo.st_size))
            row.append("{}")
            row.append(str(len(objs)))
            row.append(str(i))
            row.append('{"name":"rect","x":%s,"y":%s,"width":%s,"height":%s}'%(str(obj[2]),str(obj[3]),str(obj[4]-obj[2]),str(obj[5]-obj[3])))
            row.append('{}')
            writer.writerow(row)
            i+=1
        
        
csvFile.close()
```
